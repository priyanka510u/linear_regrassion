```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df = pd.read_csv('car_crashes.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
lab  = LabelEncoder()
```

```python
df['abbrev'] = lab.fit_transform(df['abbrev'])
```

```python
plt.bar(df['abbrev'],df['total'])
```

```python
x = df.drop(columns=['total'])
y = df['total']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2_score(y_test,y_pred)
```

```python
plt.figure(figsize=(14,6))

plt.subplots(1,1)
plt.scatter(x_test['speeding'],y_test,color='black')
plt.plot(x_test['speeding'],y_pred,'r')
plt.xlabel('speeding')
plt.ylabel('total acciedent')

plt.subplots(1,1)
plt.scatter(x_test['alcohol'],y_test,color='black')
plt.plot(x_test['alcohol'],y_pred,'r')
plt.xlabel('alcohol')
plt.ylabel('total acciedent')

plt.subplots(1,1)
plt.scatter(x_test['not_distracted'],y_test,color='black')
plt.plot(x_test['not_distracted'],y_pred,'r')
plt.xlabel('not_distracted')
plt.ylabel('total acciedent')



plt.subplots(1,1)
plt.scatter(x_test['no_previous'],y_test,color='black')
plt.plot(x_test['no_previous'],y_pred,'r')
plt.xlabel('no_previous')
plt.ylabel('total acciedent')

plt.subplots(1,1)
plt.scatter(x_test['ins_premium'],y_test,color='black')
plt.plot(x_test['ins_premium'],y_pred,'r')
plt.xlabel('ins_premium')
plt.ylabel('total acciedent')

plt.subplots(1,1)
plt.scatter(x_test['ins_losses'],y_test,color='black')
plt.plot(x_test['ins_losses'],y_pred,'r')
plt.xlabel('ins_losses')
plt.ylabel('total acciedent')

plt.subplots(1,1)
plt.scatter(x_test['abbrev'],y_test,color='black')
plt.plot(x_test['abbrev'],y_pred,'r')
plt.xlabel('abbrev')
plt.ylabel('total acciedent')
```
